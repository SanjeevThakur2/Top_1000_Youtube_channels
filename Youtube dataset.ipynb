### Dropping The Null Values and Replacing the Null Value'

```python
# dropping null value
subs.drop(subs_0.index,inplace = True)
```

```python
# examing changes
subs.shape
```

```python
subs.dtypes
```

```python
# replacing comma from numerical data of Subscribers
subs.Subscribers = subs["Subscribers"].str.replace(",","")
```

```python
# replacing Video Views
subs["Video Views"] = subs["Video Views"].str.replace(",","")
```

```python
# replacing comma of Video Count
subs["Video Count"] = subs["Video Count"].str.replace(",","")
```

```python
#examing data type
subs.dtypes
```

```python
# change the data type
subs = subs.astype({"Subscribers":'int64', "Video Views":'int64',"Video Count":'int64'})
```
